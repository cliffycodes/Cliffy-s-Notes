#### 0. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Import Data

```python
file_path = r'B:\jupyter\Projects\Cheat Sheets\Sci Kit Learn Data\heart-disease.csv'
heart_disease = pd.read_csv(file_path)
heart_disease
```

#### 2. Set the feature matrix and target variable to predict

```python
# X : Set the features matrix 
X = heart_disease.drop("target",axis=1)

# Y : Set the label (Variable to predict)
Y = heart_disease["target"]
```

#### 3. Choose the right model and hyperparameters

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

# Default Hyperparameters
clf.get_params()
```

#### 4. Fit the model to the training data

```python
from sklearn.model_selection import train_test_split

#Split our data from X and Y to leave some parts for training and testing
#0.2 = 20% data will be testing while 80% willbe used for training
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2)
```

```python
#Find the patterns from the training data
clf.fit(X_train,y_train)
```

```python
# Make a prediction to test
# Must have the same dimension as the features/columns we have fitted to the model
y_preds = clf.predict(X_test)
y_preds
```

```python
X_test
```

```python
X_test["target"] = y_preds
X_test
```

```python
y_test
```

### 5. Evaluate the model

#### Accuracy Metric

```python
#1.0 is 100% accurate
clf.score(X_train,y_train)
```

```python
#85% accuracy (Model has never seen this data nor the labels)
clf.score(X_test,y_test)
```

#### Classification Matrix
Compares the test labels and prediction labels

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print (classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

### 6. Improve Model

```python
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print (f"Trying model with {i} estimators....")
    clf = RandomForestClassifier(n_estimators=i)
    clf = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"Model accuracy on test set: {clf.score(X_test,y_test) * 100:.2f}%")
    print("")
```

### 7. Save the model and load it

```python
import pickle

pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```
